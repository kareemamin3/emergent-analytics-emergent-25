```python
import numpy as np
import pandas as pd
from numpy import cov
from scipy.stats import pearsonr
import datetime
```

```python
gdp_raw = pd.read_excel('/project_data/data_asset/sun/economy/Eurostat_GDP.xls')
prod_raw = pd.read_excel('/project_data/data_asset/sun/economy/Eurostat_ProductionIndex.xlsx','Sheet 1')
#print(prod_raw.sheet_names)
retail_raw = pd.read_excel('/project_data/data_asset/sun/economy/Eurostat_Table_Retail-TO.xlsx','Sheet 1')
```

Extract the Production Index

```python
print(prod_raw)
```

```python
prod = pd.DataFrame()
prod = prod_raw[8:47]
prod_2 = pd.DataFrame()
#prod_2.append(prod[0,1])

for i in range(2,26, 2):
    prod.drop(['Unnamed: '+str(i)], axis = 1,inplace=True)
prod.drop([9], axis = 0,inplace=True)
prod = prod.transpose()

new_header = prod.iloc[0] #grab the first row for the header
prod = prod[1:] #take the data less the header row
prod.columns = new_header #set the header row as the df header

prod.reset_index(drop=True,inplace=True)
prod.replace(':',np.nan,inplace=True)

print(prod)
```

Extract retail index

```python
print(retail_raw)
```

```python
retail = pd.DataFrame()
retail = retail_raw[8:47]

for i in range(2,26, 2):
    retail.drop(['Unnamed: '+str(i)], axis = 1,inplace=True)
retail.drop([9], axis = 0,inplace=True)
retail = retail.transpose()

new_header = retail.iloc[0] #grab the first row for the header
retail = retail[1:] #take the data less the header row
retail.columns = new_header #set the header row as the df header

retail.reset_index(drop=True,inplace=True)
retail.replace(':',np.nan,inplace=True)

print(retail)
```

**Calculate covariance & correlation**

```python
## for France
prod_fr = pd.to_numeric(prod['France'])
#prod_fr.to_numeric(s, errors='coerce')

retail_fr = pd.to_numeric(retail['France'])

covariance_fr = cov(prod_fr, retail_fr)
print(covariance_fr)
corr_fr, _ = pearsonr(prod_fr, retail_fr)
print(corr_fr)
```

```python
print(type(retail.columns))
```

```python
correlation = dict()
for index in range(2,37):
    #print('Column Number : ', index)
    # Select column by index position using iloc[]
    prod_col = pd.to_numeric(prod.iloc[: , index])
    retail_col = pd.to_numeric(retail.iloc[: , index])
    
    if not(prod_col.isnull().values.any() or retail_col.isnull().values.any()):
        #print(prod_col.values)
        corr, _ = pearsonr(prod_col, retail_col)
        #print(corr)
        correlation[prod.columns.values[index]] = corr
    #print('Column Contents : ', columnSeriesObj.values)
print(correlation)
```
