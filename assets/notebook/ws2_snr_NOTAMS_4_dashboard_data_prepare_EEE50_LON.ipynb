```python
import pandas as pd
import csv
```

```python
df = pd.read_csv("/project_data/data_asset/ws2/notams/valid_airport_notams_with_quarantine_20200717.csv")
df_airspace = pd.read_csv("/project_data/data_asset/ws2/notams/valid_airspace_notams_with_quarantine_20200717.csv")
```

```python
df.head()
```

```python
len(df)
```

```python
dash_dict = dict()
dash_dict['Closed'] = []
dash_dict['airportCode'] = []
dash_dict['airportName'] = []
dash_dict['cityName'] = []
dash_dict['countryCode'] = []
dash_dict['countryName'] = []
dash_dict['latitude'] = []
dash_dict['longitude'] = []
dash_dict['cleaned_message'] = []
dash_dict['quarantine_text'] = []
dash_dict['quarantine_days'] = []
for ac, r in df.groupby(['airportCode']):
    dash_dict['Closed'].append(r['Closed'].unique()[0])
    dash_dict['airportCode'].append(r['airportCode'].unique()[0])
    dash_dict['airportName'].append(r['airportName'].unique()[0])
    dash_dict['cityName'].append(r['cityName'].unique()[0])
    dash_dict['countryCode'].append(r['countryCode'].unique()[0])
    dash_dict['countryName'].append(r['countryName'].unique()[0])
    dash_dict['latitude'].append(r['latitude'].unique()[0])
    dash_dict['longitude'].append(r['longitude'].unique()[0])
    
    all_cm = []
    for cm in r[~r['cleaned_message'].isna()]['cleaned_message']:
        all_cm.append(cm)
    
    dash_dict['cleaned_message'].append('. '.join(all_cm))
    
    all_qt = []
    for qt in r[~r['quarantine_text'].isna()]['quarantine_text']:
        all_qt.append(qt)
    
    dash_dict['quarantine_text'].append('. '.join(all_qt))
    
    all_qt_days = []
    for qt_ in r[~r['quarantine_days'].isna()]['quarantine_days']:
        all_qt_days.append(qt_)
    
    dash_dict['quarantine_days'].append(', '.join(all_qt_days))
```

```python
dash_df = pd.DataFrame(dash_dict)
```

```python
dash_df['Closed'] = dash_df['Closed'].astype(int)
```

```python
dash_df.to_csv("/project_data/data_asset/ws2/notams/dashboard_airport.csv",index=False,quoting=csv.QUOTE_NONNUMERIC)
```
