```python
from pixiedust.display import *
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get("https://www.tsa.gov/coronavirus/passenger-throughput")

soup = BeautifulSoup(r.text)

df = pd.read_html(soup.find_all("table")[0].encode("ascii"),header=0)[0].dropna()
df["date"] = pd.to_datetime(df.Date)
df = df.rename(columns={'Total Traveler Throughput':"travellers2020",'Total Traveler Throughput  (1 Year Ago - Same Weekday)':"travellers2019"})
```

```python
display(df)
```

```python
#ddf = df[df.date > pd.to_datetime("2020-04-01")].copy()
ddf = df.copy()
ddf.index = ddf.date
ddf.index.name = "index"
ddf.sort_index(inplace=True)
ddf["2020_relative_to_prev_week"] = ddf.travellers2020.pct_change(7)*100
xdf = ddf.resample("1W").mean()
xdf["refdate"] = xdf.index
```
